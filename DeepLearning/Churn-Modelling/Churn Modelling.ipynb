```python
import pandas as pd
```

```python
dataset = pd.read_csv("Churn_Modelling.csv")
```

```python
dataset
```

```python
dataset.columns
```

```python
y = dataset['Exited']
```

```python
X = dataset[['CreditScore','Age','Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary']]
```

```python
X.shape
```

# One Hot Encoding For Geograhy  feature

```python
Geo = pd.get_dummies(dataset['Geography'], drop_first=True)
```

```python
Geo
```

# One Hot Encoding For Gender feature

```python
Gender = pd.get_dummies(dataset['Gender'], drop_first=True)
Gender
```

```python
X = pd.concat([X,Geo,Gender], axis=1)
X.shape
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=45)
```

```python
from keras.layers import Dense
from keras.models import Sequential
model = Sequential()
```

```python
model.add(Dense(units=8, activation='relu', input_dim=11))
```

```python
model.add(Dense(units=8, activation='relu'))
```

```python
model.summary()
```

```python
model.add(Dense(units=10, activation='relu'))
```

```python
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.get_config()
```

```python
model.summary()
```

```python
from keras.optimizers import Adam
```

```python
model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=0.0001))
```

```python
model.get_weights()
```

```python
model.fit(X_train, y_train, epochs=150)
```

```python
y_test
```

```python
yhat = model.predict(X_test)
```

```python
import numpy as np
yhat= (yhat > 0.5).astype(np.float32)
```

```python
yhat
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test,yhat)
```

```python
confusion_matrix(y_test,yhat)
```
